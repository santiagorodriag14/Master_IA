### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí
promedio = 0
valor_maximo = esperanzas_de_vida[0]
valor_minimo = esperanzas_de_vida[0]
# Promedio
for i in esperanzas_de_vida:
    promedio += i

promedio = promedio/len(esperanzas_de_vida)

# Máximo
for i in esperanzas_de_vida:
    if i > valor_maximo:
        valor_maximo = i

# Mínimo
for i in esperanzas_de_vida:
    if i < valor_minimo:
        valor_minimo = i

print(f"La esperanza de vida promedio de la lista es: {promedio:.2f}")
print(f"La esperanza de vida máxima es: {valor_maximo}")
print(f"La esperanza de vida mínima es: {valor_minimo}")

````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí
def clasificar_esperanza_vida(anos):
    if anos < 60:
        return "Baja"
    elif anos>= 60 and anos <= 74.9:
        return "Media"
    elif anos >= 75:
        return "Alta"

prueba1 = clasificar_esperanza_vida(50)
prueba2 = clasificar_esperanza_vida(60)
prueba3 = clasificar_esperanza_vida(72)
prueba4 = clasificar_esperanza_vida(74.9)
prueba5 = clasificar_esperanza_vida(80)

print(f"Prueba 1: {prueba1}")
print(f"Prueba 2: {prueba2}")
print(f"Prueba 3: {prueba3}")
print(f"Prueba 4: {prueba4}")
print(f"Prueba 5: {prueba5}")
````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Tu código aquí
df_nuevo = df

media_esperanza_vida = df_nuevo['esperanza_vida'].mean()
df_nuevo['esperanza_vida'] = df_nuevo['esperanza_vida'].fillna(media_esperanza_vida)

media_gasto_salud_usd = df_nuevo['gasto_salud_usd'].mean()
df_nuevo['gasto_salud_usd'] = df_nuevo['gasto_salud_usd'].fillna(media_gasto_salud_usd)

media_mortalidad_infantil = df_nuevo['mortalidad_infantil'].mean()
df_nuevo['mortalidad_infantil'] = df_nuevo['mortalidad_infantil'].fillna(media_mortalidad_infantil)

media_medicos_por_1000 = df_nuevo['medicos_por_1000'].mean()
df_nuevo['medicos_por_1000'] = df_nuevo['medicos_por_1000'].fillna(media_medicos_por_1000)

media_obesidad_pct = df_nuevo['obesidad_pct'].mean()
df_nuevo['obesidad_pct'] = df_nuevo['obesidad_pct'].fillna(media_obesidad_pct)

media_diabetes_pct = df_nuevo['diabetes_pct'].mean()
df_nuevo['diabetes_pct'] = df_nuevo['diabetes_pct'].fillna(media_diabetes_pct)

media_tabaquismo_pct = df_nuevo['tabaquismo_pct'].mean()
df_nuevo['tabaquismo_pct'] = df_nuevo['tabaquismo_pct'].fillna(media_tabaquismo_pct)

media_camas_hospital_por_1000 = df_nuevo['camas_hospital_por_1000'].mean()
df_nuevo['camas_hospital_por_1000'] = df_nuevo['camas_hospital_por_1000'].fillna(media_camas_hospital_por_1000)

media_vacunacion_pct = df_nuevo['vacunacion_pct'].mean()
df_nuevo['vacunacion_pct'] = df_nuevo['vacunacion_pct'].fillna(media_vacunacion_pct)

df_nuevo.isnull().sum()
````

**Responde aquí** (doble clic para editar):

*¿Por qué usamos la media para rellenar nulos en vez de eliminar las filas?*

> Tu respuesta: Se utliza el promedio porqué este valor no afecta de forma significativa la distribución de los datos, por ende, es mejor utilizar el promedio en vez de eliminarlos, obteniendo así un mayor volumen de datos.
