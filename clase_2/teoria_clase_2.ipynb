## Listas y diccionarios
Listas: 
Colección ordenada, implementada como un array dinámico, su acceso lo tiene por indices.



```python
#Lista de estudiantes
estudiantes = ["Juan", "Eduardo", "Santiago"]
print(estudiantes[2])
print(estudiantes[-1])
print(len(estudiantes))

estudiantes.append("Gilberto")
estudiantes
```

# Diccionarios:

Son pares de clave - valor



```python
#Registro de estudiantes
estudiante = {
    "nombre" : "Santiago Rodriguez",
    "edad" : 20,
    "correo" : "santiago@gmail.com",
    "activo" : True,
    "promedio" : 4.9}

print(f"Estudiante: {estudiante["nombre"]}")
estudiante["activo"] = False
print(f"Activo: {estudiante['activo']}")
```
